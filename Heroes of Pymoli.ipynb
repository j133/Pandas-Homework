```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Load in file
json_path = "../Homework/purchase_data.json"
```

```python
# Read and display the json file with Pandas
purchase_df = pd.read_json(json_path)
purchase_df.head()
```

```python
# Calculate total number of players
total_players = len(purchase_df)
total_players

#Create a table 
total_table = pd.DataFrame({"Total Players": [total_players]})
total_table = total_table[["Total Players"]]
total_table = total_table.round(2)
total_table
```

```python
# Calculate number of Unique items
unique = purchase_df["Item ID"].unique()
total_unique = len(unique)
total_unique

# Average Purchase Price
avg_price = purchase_df["Price"].mean()
avg_price

# Total Number of Purchases
total_purchase = purchase_df["Item ID"].count()
total_purchase

# Total Revenue
total_revenue = (avg_price)*(total_purchase)
total_revenue

# Create a table for purchasing analysis total
analysis_table = pd.DataFrame({"Number of Unique Items": [total_unique], "Average Price": [avg_price],"Number of Purchases": [total_purchase],"Total Revenue": [total_revenue]})
analysis_table = analysis_table[["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"]]
analysis_table = analysis_table.round(2)
analysis_table
```

```python
# Gender Demogrphaics
```

```python
# Count of Male Players
count_male = purchase_df["Gender"].value_counts()["Male"]
count_male

# Count of Female Players
count_female = purchase_df["Gender"].value_counts()["Female"]
count_female

# Count of Other/Non-Disclosed
count_other = total_players - count_male - count_female
count_other

# Percentage of Male Players
percent_male = count_male / total_players *100
percent_male

# Percentage of Female Players
percent_female = count_female / total_players *100
percent_female

# Percentage of others
percent_other = count_other / total_players *100
percent_other

# Create a table
gender_demo = pd.DataFrame({"Count": [count_male, count_female, count_other], "Percentage of Players": [percent_male, percent_female, percent_other]})
gender_demo.rename(index={0:'Male', 1:'Female', 2:'Other/Non-Disclosed'}, inplace=True)
gender_demo
```

```python
# Purchasing Analysis by Gender
```

```python
# Purchase count by males
count = purchase_df["Gender"].value_counts()["Male"]
count

count_male = purchase_df.loc[purchase_df["Gender"]=="Male"]
count_male

# Average purchase price for males
avg_male = count_male["Price"].mean()
avg_male

# Total Number of Purchases for males
male_total = count_male["Price"].sum()
male_total

# Normalized Totals for males
male_normal = male_total/count
male_normal

#Purchase count by females
f_count = purchase_df["Gender"].value_counts()["Female"]
f_count

count_female = purchase_df.loc[purchase_df["Gender"]=="Female"]
count_female

# Average purchase price for females
avg_female = count_female["Price"].mean()
avg_female

# Total Number of purches for females
female_total = count_female["Price"].sum()
female_total

# Normalized totals for females
female_normal = female_total/f_count
female_normal

# Other count


other_count = purchase_df.loc[purchase_df["Gender"]=="Other/Non-Disclosed"]
other_count

avg_other = other_count["Price"].mean()
avg_other

other_total = other_count["Price"].sum()
other_total

other_normal = other_total/11
other_normal

# Create Table
gender_table = pd.DataFrame({"Purchase Count": [count, f_count, count_other], "Average Purchase Price": [percent_male, percent_female, percent_other], "Total Purchase Value":[male_total, female_total, other_total], "Normalized Totals":[male_normal, female_normal, other_normal]})
gender_table.rename(index={0:'Male', 1:'Female', 2:'Other/Non-Disclosed'}, inplace=True)
gender_table
```
